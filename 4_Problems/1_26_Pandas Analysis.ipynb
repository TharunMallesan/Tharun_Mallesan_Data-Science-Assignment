```python
!pip install -U datasets
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

## 🟩 Calculate Median Salary (1.26.2)

```python
df["job_posted_date"]=pd.to_datetime(df['job_posted_date'])
df.groupby('job_title_short')['salary_year_avg'].median()
```

```python
df.info()
```

---

## 🟨 Average Hourly Salary by Job Title (1.26.3)

```python
df.groupby('job_title_short')['salary_hour_avg'].mean().sort_values(ascending=False)
```

## 🟨 Job with Lowest Salary (1.26.4)

## 🟨 Job Count by Country (1.26.5)

```python
# prompt: Job Count by Country

df_jobs_country = df.groupby('job_country').size().sort_values(ascending=False)
df_jobs_country
```
